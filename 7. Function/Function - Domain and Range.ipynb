```python
import pylab
import math
import numpy as np
```

# Function - Domain and Range

Problem 
Graph of the curve y = sin(x) where the domain is
$$\{x \mid 0 \le x \le 2\pi \}$$


```python
x = np.linspace(0, 2* math.pi, 100)  # 100 evenly-spaced values from 0 to 2 PI
y = np.sin(x)

pylab.plot(x, y)
pylab.title("sine wave")
```

So the range is 
$$\{x \mid -1 \le x \le 1 \}$$
