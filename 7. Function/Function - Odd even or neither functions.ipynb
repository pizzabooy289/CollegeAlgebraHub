```python
import pylab
import math
import numpy as np
```

# Function - Odd even or neither functions

## Even Function

```python
x = np.linspace(-2, 2, 100)  # 100 evenly-spaced values from -2 to 2 
y = x**2+1

pylab.plot(x, y)
pylab.show()
```

## Odd Function

```python
x = np.linspace(-2, 2, 100)  # 100 evenly-spaced values from -2 to 2 
y = x**3-x

pylab.plot(x, y)
pylab.show()
```
